## Train - test splitting 

```python
import numpy as np
def split_train_test(data,test_ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_set_size= int(len(data)*test_ratio)
    test_indices= shuffled[:test_set_size]
    train_indices=shuffled[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

## scikit-learn Design

##### primarily , three types of objects
##### 1.Estimators :- It estimates same Parameter based on dataset .eg, imputer , it has a fit method and transform method . fit method :- fits the datasets and  calulates internal paramters
##### 2Transformers :- Transformn  method takes input and returns output based on the learnings from fit().it also has aconvenience function called fit_transform() which fits and then transforms
##### 3.predictors :- linear Regression model is an example of predictor,fit() predic()It are two common functions.It also  given score() function which wll evalute the predictions.


## feature Scalling

### creating a pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    # ................ add as many as you want in your pipline
    ('std_scaler',StandardScaler()),
])
```

```python
df_num_tr = my_pipeline.fit_transform(df)
```

```python
df_num_tr.shape
```

# Evaluating the model

```python
from sklearn.metrics import  mean_squared_error
df_predictions = model.predict(df_num_tr)
mseb = mean_squared_error(df_labels,df_predictions)
rmse = np.sqrt(mseb)
```

```python
rmse
```

## using better evaluation technique - Cross Valiodation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,df_num_tr,df_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def  print_scores(scores):
    print("Scores:",scores)
    print("Mean:",scores.mean())
    print("Standard deviation:",scores.std())
```

```python
print_scores(rmse_scores)
```
